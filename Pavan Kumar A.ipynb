## 3: Data Preparation

### Creating Dummy variables

```python
## The Dummy variables for month, season, weathersit, weekday are created below

months=pd.get_dummies(b.month,drop_first=True)
weekdays=pd.get_dummies(b.weekday,drop_first=True)
weather_sit=pd.get_dummies(b.weathersit,drop_first=True)
seasons=pd.get_dummies(b.season,drop_first=True)
```

```python
# Combine results to the original bike dataframe

b=pd.concat([months,weekdays,weather_sit,seasons,b],axis=1)
b.head()
```

```python
# Dropping 'season','month','weekday','weathersit' 

b.drop(['season','month','weekday','weathersit'], axis = 1, inplace = True)
b.head()
```

## 4: Splitting the Data into Training and Testing Sets

```python
# this is specified so that the train and test data set always have the same rows, respectively

btrain, btest = train_test_split(b, train_size = 0.7, random_state = 100)
```

```python
# The Rows and columns after split

print(btrain.shape)
print(btest.shape)
```

# Rescaling the features
It is important to have all the variables on the same scale for the model to be easily interpretable. We can use standardization or normalization so that the units of the coefficients obtained are all on the same scale.
Two common ways of rescaling:

   - Min-Max scaling (Normalisation):Between 0 and 1

   - Standardisation :mean-0, sigma-1



```python
#Instantiate an object
scaler = MinMaxScaler()

#Create a list of numeric variables
num_vars=['temp','humidity','windspeed','count']

#Fit on data
btrain[num_vars] = scaler.fit_transform(btrain[num_vars])
btrain.head()
```

```python
#Checking numeric variables(min and max) after scaling
btrain.describe()
```

All the numeric variables are now mapped between 0 and 1

## Multicollinearity
Assumes that predictors are not correlated with each other. If there is correlation among the predictors

```python
# We will check the correlation coefficients to see which variables are highly correlated after scaling

plt.figure(figsize=(25, 20))
sns.heatmap(btrain.corr(),cmap='YlGnBu',annot = True)
plt.show()
```

## Observation
From the heatmap we can see temp, year is  correlated to the 'count' variable. Even in August, September months we can see the counts are little high.

```python
#Divide the data into X and y
y_train = btrain.pop('count')
X_train = btrain
```

```python
# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
#List of variables selected
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
#Columns where RFE support is True
col = X_train.columns[rfe.support_]
col
```

```python
#Columns where RFE support is False
X_train.columns[~rfe.support_]
```

```python
# Building using stats model
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
# Running the linear model 
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
print(lm.summary())
```

## Checking VIF

```python
#Drop the constant term B0
X_train_rfe = X_train_rfe.drop(['const'], axis=1)
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

January is insignificant in presence of other variables due to high p-value and low VIF; can be dropped

```python
#Drop January
X_train_new1 = X_train_rfe.drop(["Jan"], axis = 1)
```

#### The VIFs and p-values both are within an acceptable range. 




**- The p-value for each term tests the null hypothesis that the coefficient is equal to zero (no effect). A low p-value (< 0.05) indicates that we can reject the null hypothesis.**

**-  A rule of thumb commonly used in practice is if a VIF is > 10, you have high multicollinearity. In our case, with values less than 5, we are in good shape, and can proceed with our regression**

**- R-squared measures the strength of the relationship between your model and the dependent variable on a convenient 0 – 100% scale. And we have the R-square value of 0.826 or 82.6%**

**- The adjusted R-squared adjusts for the number of terms in the model. And we got it around 0.82 or 82%**

#### So our model is lm5 which is obtained by removing January, windspeed, holiday, July and humidity variables from the RFE support columns
